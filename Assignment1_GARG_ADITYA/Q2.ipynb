solve the following system : 

$$y' = (1+4x) \sqrt{y}$$

```python
from math import * 
from ode_solver import *

def f(x, y):
    
    return (1+(4*x))*sqrt(y)

x, y = euler(f, 0, 5, 25, 1)
x_, y_ = rk4(f, 0, 5, 25, 1)
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

plt.figure(figsize=(20, 6))
plt.xlabel('x')
plt.ylabel('y')

plt.plot(np.array(x), (np.array(x)+((np.array(x)**2)/2+1))**2, 'r--', label ='analytical')
plt.plot(np.array(x_), np.array(y_), 'b', label = 'ERK4')
plt.plot(np.array(x), np.array(y), 'g--', label = 'euler')
plt.legend()
```
