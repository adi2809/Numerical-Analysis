# Euler Method 

<h4>pseudo-code :</h4>

- start : 

- define the function $f(x,y)$

- read the values of initial condition at $x = x_0$, $y = y_0$, the stepsize and calculation point $x_n$
   
- calculate number of steps $n = \frac{x_n - x_0}{h}$

- initialise the loop variable with 0. 

- for i in range(n): 
      
$$y_n = y_0 + h *  f(x_0 + i \cdot h, y_0)$$
      
$$y_0 = y_n$$
      
$$i = i + 1$$

   

- display $y_n$ as result

- stop

the function has been implemented in the "ode_solver.py"

```python
from ode_solver import * 
```

```python
##trial run##
def f(x, y):
    
    return -(x*y)

x, y = euler(f, 0, 3, 300, 1)
```

```python
import matplotlib.pyplot as plt
import numpy as np 

plt.figure(figsize=(10, 3))
plt.xlabel('x')
plt.ylabel('y')



plt.plot(np.array(x), np.exp(-(np.array(x)**2)), 'r', label ='analytical')
plt.plot(np.array(x), np.array(y), 'b', label = 'euler')

plt.legend()
```

# Fourth order Explicit RK method:

<h4>pseudo-code :</h4>


- start

- define the function f(x,y)

- Read values of initial condition ($x_0$ and $y_0$), step size ($h$) and calculation point ($x_n$)
   
- calculate the value of number of steps $n = \frac{(x_n-x_0}{h}$

- intialise i = 0 

- loop for i < n
      
$$k_1 = h f(x_0, y_0)$$
      
$$k_2 = h  f(x_0+\frac{h}{2}, y_0+\frac{k_1}{2})$$
      
$$k_3 = h  f(x_0+\frac{h}{2}, y_0+\frac{k_2}{2})$$
      
$$k4 = h f(x_0+h, y_0+k_3)$$
      
$$k = \frac{(k_1+2k_2+2k_3+k_4)}{6}$$
      
$$y_n = y_0 + k$$
      
$$i = i + 1$$
$$x_0 = x_0 + h$$
$$y_0 = y_n$$
      
  


- return the ouputs $y_n$

- stop

the function has been implemented in the "ode_solver.py"

```python
##trial run##
def f(x, y):
    
    return -(x*y)

x_, y_ = rk4(f, 0, 3, 300, 1)
```

```python
plt.figure(figsize=(10, 3))
plt.xlabel('x')
plt.ylabel('y')

# plt.xscale('log')
# plt.yscale('log')

plt.plot(np.array(x_), np.exp(-(np.array(x_)**2)), 'r', label ='analytical')
plt.plot(np.array(x_), np.array(y_), 'b', label = 'ERK4')
plt.plot(np.array(x), np.array(y), 'g--', label = 'euler')
plt.legend()
```
