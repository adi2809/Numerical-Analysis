```python
from math import * 

def f(x, y):
    """k = 0.06"""
    return -0.06*(sqrt(y))

print('---------------------------------------')
x, y = euler(f, 0, 57, 275 , 3)

print('---------------------------------------')
x_, y_ = rk4(f, 0, 57.6, 288, 3)
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

#rk4 

plt.figure(figsize=(20, 10))
plt.xlabel('x')
plt.ylabel('y')



plt.plot(np.array(x_), np.array(y_), 'b', label = 'ERK4')
```

```python
#euler
plt.figure(figsize=(20, 10))
plt.plot(np.array(x), np.array(y), 'g--', label = 'euler')
```

from the euler method time for water level to drop to zero is around 57 seconds. 

from the rk4 method time for the water level to drop to zero is 57.6 seconds.
